```python
import torch
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/Users/emrecanacikgoz/Desktop/DSC_1474_x1.png')
plt.imshow(image);
print(f"Image dim: {image.shape}")
```

```python
operatedImage = np.float32(image)
```

```python
dst = cv2.cornerHarris(operatedImage[:,:,0], 3, 3, 0.04)
```

```python
thres1 = 0.01*dst.max()
thres2 = -0.001*dst.max()
```

```python
map_corner = np.float32(dst > thres1)
map_edge = np.float32(dst < thres2)
map_flat = np.float32((dst > thres2) & (dst < thres1))
```

```python
y_corner = np.transpose(np.expand_dims(map_corner, axis=0), (1,2,0))
y_edge = np.transpose(np.expand_dims(map_edge, axis=0), (1,2,0))
y_flat = np.transpose(np.expand_dims(map_flat, axis=0), (1,2,0))
```

```python
plt.imshow(y_corner, 'gray');
```

```python
plt.imshow(y_edge, 'gray');
```
